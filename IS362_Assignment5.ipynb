<b style="color: #FF0000;"> Q1. What is the northernmost airport in the United States? </b>

<i style="color: #006600;"> We begin by importing pandas and then continue by reading the data from the file provided by the Professor. 
The data will alow us to begin answering the question for this week's task. </i>

```python
import pandas as pd # Importing pandas Module as Variable Called pd
airports = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv') #reading the data
airports.sort_values(by = 'lat', ascending = False).head(5) #sorts the data by latitude & displays top 5 results
```

<p style="color: #006600;">According to the data provided to us we can see the northernmost airport in the United States is
<b>Dilliant Hopkins Airport </b> in New Hampshire. After researching online, I found this to be incorrect. According to [Maps of the World](https://www.mapsofworld.com/usa/states/alaska/alaska-airports.html), the northernmost airport in the US is <b><i> Wiley Post Will Rogers Memorial Airport </i></b> located in Alaska (71°17′08″N 156°45′58″W).</p>

<b> Q2. What is the easternmost airport in the United States? </b>

<i> We will continue to use the same .csv data file as for Q1 and extract the top 5 easternmost airport in the US. </i>

```python
airports.sort_values(by = 'lon', ascending = False).head(5) #sorts the data by longitude & displays top 5 results
```

<p style="color: #006600;">According to the data provided to us we can see the easternmost airport in the United States is <b> Eareckson As Airport </b> in Alaska; which we can confirm is incorrect as Alaska is on western part of the US. According to [Airline Pilot Control](https://www.airlinepilotforums.com/hangar-talk/53045-most-extreme-airports-us.html), the northernmost airport in the US is <b><i> Eastport Municipal Airport </i></b> located in Maine (44D54.61'N 67D00.76W).</p>

<b style="color: #FF0000;"> Q3. On February 12th, 2013, which New York area airport had the windiest weather? </b>

<i style="color: #006600;"> For this question we will read the data from the <i> weather.csv </i> file to determine weather related data. </i>

```python
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
#reads data from weather .csv

weather[(weather['year'] == 2013)
        & (weather['month'] == 2)
        & (weather['day'] == 12)].sort_values(by = ['wind_gust'], ascending=False).head(5) 
#filters February 12, 2013 in descending order by wind gust
```

<p style="color: #006600;">According to the data provided by the Professor, it appears that <b> Newark Airport </b> has the windest weather; but this is not a NY airport, instead a NJ airport. The high wind gust in New Jersey may have occurred due to the 2013 [Nor'ester](https://www.ncdc.noaa.gov/sotc/national/201302). The correct answer is <b><i> LGA </i></b> located in the NYC area.</p>
